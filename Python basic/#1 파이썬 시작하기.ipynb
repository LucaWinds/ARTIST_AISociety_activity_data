> **용어 정리**

문장 : 실행 가능한 최소 단위 <br>
프로그램 : 문장이 모임 <br>
표현식 : 값을 가지는 식 <br>
```
# 표현식 예시
1 + 3
12 // 2
"Python"
```
키워드 : 미리 약속해놓은 특별한 단어
```
# 키워드 예시
if
False
class

# False는 키워드지만,
# false는 키워드 아님.
```
식별자 : 이름 붙일 때 사용하는 단어
```
#가능한 단어
alpha
alpha10
al_pha
Alpha
ALPHA

#불가능한 단어
break    #예약어
10alpha  #숫자 시작
al pha   #공백 존재
```
<font size="2em" color="gray">
snake_case : 언더바를 기호 중간에 <br>
일반적으로 함수 또는 변수에 사용 <br>
CamelCase : 단어들의 첫 글자를 대문자로 <br>
일반적으로 클래스에 사용 <br>
</font>

```

```
연산자 : 값과 값 사이에 무언가 기능을 가짐
```
#아래 예시에서는 +가 연산자
2 + 3
```
리터럴 : 값 자체
```

```
