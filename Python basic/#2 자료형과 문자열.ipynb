

> **문자열 다루기**

* 문자열 연결 연산자

```
"문자열" + "문자열"
# -> 문자열 연결
```

* 문자열 반복 연산자

```
"문자열" * 숫자
# -> 문자열 반복
```

* 문자열 선택 연산자 (인덱싱)

  문자열 내부의 문자 하나를 인덱스로 지정 <br>
  인덱스는 0부터 시작 <br>
  인덱스가 음수인 경우 뒤에서부터 센다


```
str[index]
# -> index 번째 문자 반환
```

* 문자열 범위 선택 연산자 (슬라이싱)

  문자열의 특정 범위를 선택 <br>
  값 생략을 하는 경우도 있다. <br>
  앞의 값 생략 : 0번째부터 뒤의 숫자 앞의 문자까지 <br>
  뒤의 값 생략 : n번째부터 끝의 문자까지

```
str[start:end]
# -> start부터 end 앞의 문자까지
```

> **문자열 포매팅** 

%를 이용하여 문자열 안에 어떤 값을 삽입하는 방법

<font size="2em" color="gray">
%s : 문자열(string)<br>
%c : 문자 1개(character)<br>
%d : 정수(integer)<br>
%f : 부동소수(floating-point)<br>
%o : 8진수<br>
%x : 16진수<br>
%% : 리터럴 % (% 자체)<br>
</font>

````python
print("I eat %d apples." % 3)

print("I eat %s apples." % "three")

number = 3
print("I eat %d apples." % number)

number2 = "three"
print("I eat %d apples, and I eat %s apples" % (number, number2))
````

````python
#오른쪽 정렬
print("%10s" % "hi")

#왼쪽 정렬
print("%-10sjane." % 'hi')
````

````python
# 소수점 표현하기
print("%0.4f" % 3.42134234)
print("%10.4f" % 3.42134234)
````



> **format 함수를 이용한 포매팅**

format 함수를 이용하여 인덱스를 통한 더 직관적인 대입이 가능하다

````python
# 사용법은 거의 동일하지만, 형식 지정이 필요 없음
# 인덱스는 항상 0부터 시작
print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))

print("I eat {0} apples".format(number))
print("I ate {0} apples. so I was sick for {1} days.".format(number, number2))

# 인덱스가 아닌 변수 이름으로도 지정 가능
print("I ate {number} apples. so I was sick for {day} days.".format(number=10, day=3))

# 섞어도 됨
print("I ate {0} apples. so I was sick for {day} days.".format(10, day=3))
````

````python
# 이 외에도 수많은 문자열 함수가 있습니다.

# 소문자를 대문자로 바꾸기
a = "hi"
print(a.upper())

# 대문자를 소문자로 바꾸기기
a = "HI"
print(a.lower())

# 왼쪽 공백 지우기
a = " hi "
print(a.lstrip())

# 오른쪽 공백 지우기
a = " hi "
print(a.rstrip())

# 양쪽 공백 지우기
a = " hi "
print(a.strip())

# 문자열 바꾸기
a = "Life is too short"
print(a.replace("Life", "Your leg"))
````
