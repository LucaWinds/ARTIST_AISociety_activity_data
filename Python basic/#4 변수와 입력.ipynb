# **변수와 입력**



> **변수**

값을 저장할 때 사용하는 식별자<br>
자료형에 구애받지 않음<br>

````python
#변수에 값 할당
pi = 3.14159265

#변수를 이용해 연산하기
print(pi**(0.5))
print(pi-2)
print(pi+pi)
````

````python
pi = 3.14159265
r = 10

print("원주율 =", pi)
print("반지름 =", r)
print("원의 둘레 =", 2*pi*r)
print("원의 넓이 =", pi*r*r)
````



> **복합 대입 연산자**

```
(변수) (기본연산자)= (숫자)

#이는 다음과 같습니다
(변수) = (변수) (기본연산자) (숫자)
```



````python
a = 100

a*=20
print(a)

# 문자열에도 할 수 있다
b = "안녕하세요"
b += "!"
print(b)

# 문자열에도 할 수 있다
b = "안녕하세요"
b *= 5
print(b)
````



> **사용자 입력**

`input` 함수를 통해 입력받는다. <br>
입력된 값은 무조건 "문자열"이다.

````python
string = input("입력해보세요 : ")
print(string)
print(type(string))
````



> **캐스트**

입력받은 값은 숫자 연산으로 활용하기 위해 문자열을 숫자로 변환.

````python
string = input("숫자를 입력하세요 : ")
strtoint = int(string)

print(string, type(string))
print(strtoint, type(strtoint))

# 정수가 아닌걸 바꿀려고 하면 ValueError 출력
````

````python
string = input("실수를 입력하세요 : ")
strtofloat = float(string)

print(string, type(string))
print(strtofloat, type(strtofloat))

# 실수가 아닌걸 바꿀려고 하면 ValueError 출력
````

````python
# 숫자를 문자열로 바꿀 수도 있음
num = 10
numtostr = str(num)

print(num, type(num))
print(numtostr, type(numtostr))
````
