# **불 자료형**

> Boolean

True, False로 표현. <br>
<font size="2em" color="gray">
(C++) true, false 아님. <br>
(C) 1, 0 아님. <br>
</font>

> 비교 연산자

==, !=, <, >, <=, >=

> 논리연산자

not, and, or <br>
<font size="2em" color="gray">
(C/C++) !, &&, || 아님. <br>
</font>

````python
# 문자와 불 자료형

print("가방" == "가방")
print("가방" != "하마")
print("가방" < "하마")
print("가방" > "하마")
````

````python
# 논리연산자

print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)
````

````python
number = int(input("정수 입력 : "))

if number > 0:
  print("양수입니다")

if number < 0:
  print("음수입니다")

if number == 0:
  print("0입니다")
````

> datetime

현재 날짜와 시간을 구하는 기능을 가진 모듈

````python
# 모듈 불러오기
import datetime

# 현재 날짜, 시간 구하기
now = datetime.datetime.now()

# 출력하기
print(f"{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초")

# 오전/ 오후 구분
if now.hour < 12:
  print(f"현재 시각은 {now.hour}시로 오전입니다.")
if now.hour >= 12:
  print(f"현재 시각은 {now.hour}시로 오후입니다.")

# 계절 구분
if 3<=now.month<=5:
  print(f"이번 달은 {now.month}월로 봄입니다.")
if 6<=now.month<=8:
  print(f"이번 달은 {now.month}월로 여름입니다.")
if 9<=now.month<=11:
  print(f"이번 달은 {now.month}월로 가을입니다.")
if now.month==12 or 1 <= now.month <= 2:
  print(f"이번 달은 {now.month}월로 겨울입니다.")
````

> in 연산자

안에 특정 문자가 있는지 확인하는 연산자.

````python
number = input("정수 입력 > ")
last = number[-1]

if last in "02468":
  print("짝수입니다.")
if last in "13579":
  print("홀수입니다.")
````

> **else 구문**

if 조건문 뒤에 사용하여 if 조건문이 거짓일때 실행되는 부분

```
if 표현식:
  참일때
else: # 표현식 쓰지 않음!
  거짓일 때
```

````python
number = int(input("정수 입력 > "))

if number % 2 == 0:
  print("짝수입니다.")
else:
  print("홀수입니다.")
````

> **elif 구문**

3개 이상의 조건을 연결해서 사용한다. <br>
```
if 표현식:
  참일때
elif 표현식:
  그 다음 참일때 
else: # 표현식 쓰지 않음!
  거짓일 때
```

<font size="2em" color="gray">
(C/C++) else if 아님
</font>

````python
# 모듈 불러오기
import datetime

# 현재 날짜, 시간 구하기
now = datetime.datetime.now()

# 출력하기
print(f"{now.year}년 {now.month}월 {now.day}일 {now.hour}시 {now.minute}분 {now.second}초")

# 계절 구분
if 3<=now.month<=5:
  print(f"이번 달은 {now.month}월로 봄입니다.")
elif 6<=now.month<=8:
  print(f"이번 달은 {now.month}월로 여름입니다.")
elif 9<=now.month<=11:
  print(f"이번 달은 {now.month}월로 가을입니다.")
else:
  print(f"이번 달은 {now.month}월로 겨울입니다.")
````
