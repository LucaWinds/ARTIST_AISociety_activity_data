# **파일 처리**

open - read- write - **close** 순 <br>

1. w : write
2. a : append
3. r = read

```python
file = open("test.txt","w")

file.write("Hello Python Programming....!")

file.close()
```

> with 키워드

파일을 열고 닫지 않는 경우 방지

```python
with open("test.txt", "w") as file:
  file.write("파이썬 근본없다!!!")
```

```python
# 읽어오기
with open("test.txt", "r") as file:
  contents = file.read()

print(contents)
```

> CSV 형태를 이용한 데이터 입출력

```python
import random
hanguls = list("김이이이이이이박우최조황석준신상훈민동성채도영연여아안수소부보빈지석진화운복목별새연보숙양재중근")
with open("info.txt", "w") as file:
  for i in range(1000):
    name = random.choice(hanguls) + "길수"
    weight = random.randrange(40, 100)
    height = random.randrange(140,200)

    file.write(f"{name}, {weight}, {height}\n")
```

```python
with open("info.txt", "r") as file:
  for line in file:
    (name, weight, height) = line.strip().split(", ")

    # 데이터가 문제 없는지 확인
    if (not name) or (not weight) or (not height):
      continue

    bmi = int(weight)/ (int(height) ** 2) * 10000 
    res = ""
    if 25 <= bmi: res = "과체중"
    elif 18.5 <= bmi: res = "정상체중"
    else: res = "저체중"

    print('\n'.join([
        "이름:{}",
        "몸무게: {}",
        "키: {}",
        "BMI: {}",
        "결과: {}"
    ]).format(name, weight, height, bmi, res))

    print()
```
