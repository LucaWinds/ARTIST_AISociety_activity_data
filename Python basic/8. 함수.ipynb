# **함수의 기본**

코드의 집합

```
def 함수 이름(매개변수, 매개변수, ...):
  문장
```

1. 매개변수: 함수 호출 시 괄호 내부에 넣는 여러가지 자료

2. 가변 매개변수 : 매개변수를 원하는만큼 할당

```
def 함수 이름(매개변수, ..., *가변 매개변수):
  문장
```
가변 매개변수는 하나만, 매개변수 뒤에 와야 한다.

3. 리턴값: 함수의 결과

```
def 함수(매개변수)
  변수 = 초기값
  #처리
  return 변수
```

````python
def print_n_times(n, *values):
  for i in range(n):
    for value in values:
      print(value)
    print()

print_n_times(3, "안녕하세요", "즐거운", "프로그래밍")
````

````python
# 자료와 함께 리턴
def return_test():
  return 100

val = return_test()
print(val)
````

````python
# 자료 없이 리턴
def return_test():
  print("A")
  return
  print("B")

return_test()
````

````python
# 리턴하지 않음
def return_test():
  return

val = return_test()
print(val)
````

````python
# 키워드 매개변수
def test(a, b=10, c=100):
  print(a+b+c)

test(10, 20, 30)
test(a=10, b=100, c=200)
test(c=10, a=100, b=200)
test(10, c=200)
````

````python
def sum_all(start,end):
  res = 0
  for i in range(start, end+1):
    res += i
  
  return res

print(sum_all(0, 100))
````

````python
# 기본 매개변수와 키워드 매개변수를 활용해 범위의 정수를 모두 더하는 함수
def sum_all(start=0, end=100, step=1):
  res = 0
  for i in range(start, end+1, step):
    res += i
  return res

print(sum_all(0, 100, 10))
print(sum_all(end=100))
print(sum_all(end=100, step=2))
````
