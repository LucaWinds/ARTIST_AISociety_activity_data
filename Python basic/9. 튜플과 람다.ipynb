2. 람다 <br>
 함수를 매개변수로 전달하여 코드를 더 효율적으로 작성

 > 표준 함수

 ```map(함수, 리스트)```
 
 ```filter(함수, 리스트)```


````python
# lambda 활용
def call_10_times(func):
  for i in range(10):
    func()

def print_hello():
  print("안녕하세요")

call_10_times(print_hello)
````

````python
# 숏코딩의 정석, 람다
# 1단계

def power(item):
  return item*item

def under_3(item):
  return item<3

list_input_a= [1,2,3,4,5]

output_a = map(power, list_input_a)
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(under_3, list_input_a)
print("filter(under_3, output_b):", output_b)
print("filter(inder_3, output_b):", list(output_b))
````

````python
# 2단계 (람다)
power = lambda x: x*x
under_3 = lambda x: x<3

list_input_a= [1,2,3,4,5]

output_a = map(power, list_input_a)
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(under_3, list_input_a)
print("filter(under_3, output_b):", output_b)
print("filter(inder_3, output_b):", list(output_b))
````

````python
# 3단계 (인라인 람다)
list_input_a= [1,2,3,4,5]

output_a = map(lambda x: x*x, list_input_a)
print("map(power, list_input_a):", output_a)
print("map(power, list_input_a):", list(output_a))
print()

output_b = filter(lambda x: x<3, list_input_a)
print("filter(under_3, output_b):", output_b)
print("filter(inder_3, output_b):", list(output_b))
````

````python
# 4단계 (마무리)
l=[1,2,3,4,5]
x,y=map(lambda x:x*x,l),filter(lambda x:x<3,l)
print(x,list(x),y,list(y),sep="\n")
````
