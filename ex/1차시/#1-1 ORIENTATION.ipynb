> 인공지능, 머신러닝, 딥러닝


**머신 러닝, Machine Learning |** <br>
사람이 하기 어려운 작업을 대신 수행할 기계를 학습을 통해 만들어내는 일련의 작업.<br><font size="2em" color="gray">
타타 프로그래밍이 규칙, 데이터를 통해 해답을 구하는 것이라면 머신러닝은 데이터와 해답을 통해 규칙을 구하는 과정입니다. </font>
<br><br>
**딥러닝, Deep Learning |** <br>
기본적으로 머신러닝의 일종으로, 학습에 필요한 데이터를 스스로 학습할 수 있는 방법. 인공신경망 기법과 거의 동일시됨<br><font size="2em" color="gray">
딥러닝은 알고리즘 기술, 빅데이터 기술, GPU 능력 향상에 의해 그 성능이 정해집니다. </font>

> 머신러닝 학습 방법론


**지도 학습, Supervised Learning |** <br>
입력값과 결과값을 함께 주고 학습을 시키는 방법<br><font size="2em" color="gray">
과거 데이터를 기반으로 앞으로 있을 이벤트를 예측하는데 탁월합니다. 주로 분류, 회귀 문제에 사용됩니다. </font><br>
<br>
**비지도 학습, Unsupervised Learning |** <br>
입력값만 주고 학습을 시키는 방법<br><font size="2em" color="gray">
데이터를 탐색하여 내부 구조와 유사한 속성을 파악합니다. 유사한 속성의 데이터를 구분하는데에 사용됩니다. </font><br>
<br>
**강화 학습, Reinforcement Learning |** <br>
인공지능이 잘 할때 마다 보상을 주는 방법<br><font size="2em" color="gray">
로봇, 게임, 내비게이션 등에 이용되어 일정 시간 내에 예상되는 보상을 극대화할 수 있는 동작을 선택하도록 합니다. </font><br>

# **인공지능을 시작하기 전 알아야 할 것들이 있습니다**

> Python

파이썬은 엄청나게 빠른 개발 속도와 생산성, 좋은 호환성, <big>*****느려터진*****</big> 실행속도와 **기적의** 실행 파일 용량을 가진 프로그래밍 언어입니다.<br>
인공지능을 다른 언어로 못하는 건 아니지만, 파이썬의 <big>압도적인</big> 편리함을 버릴 순 없었기에 인공지능을 만들때는 보통 파이썬을 사용합니다. <br>

인공지능을 다룰때 사용하는 수많은 라이브러리는 곧~언젠가~ 다루겠습니다.


> Google Colab

Jupyter Notebook이라는 Jupyter에서 제작한 Python용 통합 개발 환경을 웹에서 사용할 수 있게 구글에서 만든 툴입니다.

Colaboratory(Colab)은 브라우저 내에서 Python 스크립트를 작성할 수 있는 툴입니다.<br>
복잡한 파이썬 라이브러리 설치가 필요하지 않고(필요할 때에는 셀에 `!pip install module`를 통해 설치할 수 있습니다),<br> *****무료로***** GPU도 쓰게 해주고, 공유도 편리하기 때문에 실습에서는 Colab을 사용할 예정입니다. <br>
Colab에서는 보시는 것과 같이 실행 코드, 이미지, HTML과 LaTeX, Markdown 문법을 제공하기 때문에 시각적으로도 보기 좋습니다!
