
<font size="2em" color="gray">
불리언 값을 인덱스로 사용하는 방법 >
</font>


```
ind = (배열에 대한 조건) 
# ex) ind = arr>5 
# 각 요소에 대해 조건을 만족하는 것은 True, 
# 그렇지 않은 것은 False로 반환

arr[ind]
# True 값은 그대로, False 값은 없애버린다.
# 이후 행으로 처리된 배열을 반환한다.
```



````python
arr = np.array([[1,2],[3,4],[5,6]])
arr[[0,1,2],[0,1,0]]
````

````python
arr = np.arange(12).reshape(3,4)
arr
````

````python
ind = arr>5
ind
````

````python
arr[ind]
````

인덱스를 반환하는 추가적인 함수 >

```
np.nonzero() 
# 0이 아닌 요소의 인덱스 반환
# 즉, 0이 아닌 요소의 인덱스를 포함하는 튜플 반환
np.transpose() 
# 축을 변환
```



````python
arr = np.array([[1,0,0],[0,2,0],[1,1,0]])

np.nonzero(arr)
````

````python
arr = np.array([[0,1],[2,3]])

np. transpose(arr)
````

즉, 비교 차원들 중 크기가 1인 차원이 있다면 다른 차원과 일치시키기 위해 차원을 늘린다.

**np.newaixs**

차원을 늘린다.



```
arr = np.array([0,1,2,3])
#배열 생성

arr[np.newaxis, :]
#차원 증가 (행 벡터로 변환)

arr[:, np.newaxis]
#차원 증가 (행 벡터로 변환)
```

이를 통해 배열끼리 산술 연산, 비교 연산이 가능하다. <br>
