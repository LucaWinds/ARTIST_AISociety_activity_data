> 선형회귀

````python
#예제 1 : 데이터 표현

#모듈 불러오기
import matplotlib.pyplot as plt

#데이터 받기 (리스트)
population_inc=[0.3,-0.78,1.26,0.03,1.11,15.17,0.24,-0.24,-0.47,-0.77,-0.37,-0.85,-0.41,-0.27,0.02,-0.76,2.66]
population_old=[12.27,14.44,11.87,18.75,17.52,9.29,16.37,19.78,19.51,12.65,14.74,10.72,21.94,12.83,15.51,17.14,14.42]

#그래프 그리기
plt.plot(population_inc,population_old,'bo') #bo : 파란색 점
plt.xlabel('Population Growth Rate')  #x축 이름
plt.ylabel('Elderly Population Rate') #y축 이름
plt.show() #그래프 보여주기
````

> 최소 제곱법

**잔차** : 데이터의 경향성을 가장 잘 설명하는 하나의 직선과의 데이터의 차이<br><br>
이러한 잔차의 제곱들을 최소화하는 알고리즘을 최소 제곱법이라 한다.<br>
최소 제곱법을 통해서 직선 y=ax+b를 구하게 되면 다음과 같다.

image.png

````python
#예제 3 : 회귀식 구하기

#모듈 불러오기
import numpy as np
import matplotlib.pyplot as plt

#데이터 받기(리스트)
x=[0.3,-0.78,1.26,0.03,1.11,0.24,-0.24,-0.47,-0.77,-0.37,-0.85,-0.41,-0.27,0.02,-0.76,2.66]
y=[12.27,14.44,11.87,18.75,17.52,16.37,19.78,19.51,12.65,14.74,10.72,21.94,12.83,15.51,17.14,14.42]

#평균 구하기
x_bar=sum(x)/len(x)
y_bar=sum(y)/len(y)

#위 식을 통해 a와 b값 계산하기
a=sum([(y-y_bar)*(x-x_bar) for y,x in list(zip(y,x))])
a/=sum([(x-x_bar)**2 for x in x])
b=y_bar-a*x_bar
print('a:',a,'b:',b)

# 선형 회귀식 리스트로
line_x=np.arange(min(x),max(x),0.01)
line_y=a*line_x+b

#회귀선 그리기
plt.plot(line_x,line_y,'r-') #r- : 빨간색 선

#그래프 그리기 (위와 동일)
plt.plot(x,y,'bo')
plt.xlabel('Population Growth Rate (%)')
plt.ylabel('Elderly Population Rate (%)')
plt.show()
````

 데이터 분석<br>
 <font size="2em" color="gray">
 회귀선을 통해 인구 증가율이 증가하면 고령 인구 비율은 감소하고, 인구 증가율이 감소하면 고령인구비율이 증가할 것이라고 예측할 수 있다.

**Tensorflow를 이용한 회귀선 구하기**

````python
#모듈 불러오기
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import random

#데이터 받기(리스트)
x=[0.3,-0.78,1.26,0.03,1.11,0.24,-0.24,-0.47,-0.77,-0.37,-0.85,-0.41,-0.27,0.02,-0.76,2.66]
y=[12.27,14.44,11.87,18.75,17.52,16.37,19.78,19.51,12.65,14.74,10.72,21.94,12.83,15.51,17.14,14.42]

#a,b를 난수로 초기화
a=tf.Variable(random.random())
b=tf.Variable(random.random())

#함수 정의 : 잔차의 제곱의 평균을 구하는 함수
def compute_loss():
   y_pred=a*x+b
   loss=tf.reduce_mean((y-y_pred)**2)
   return loss

#손실값(잔차의 제곱을 모두 더해서 평균을 낸 값)을 최소화하기
optimizer=tf.optimizers.Adam(lr=0.07)
for i in range(1000):
   optimizer.minimize(compute_loss,var_list=[a,b])
    
   if i%100==99:
       print("[",(i+1)//100,"] ",'a:',a.numpy(),'b:',b.numpy(),'loss:',compute_loss().numpy())

#선형 회귀식 리스트로
line_x=np.arange(min(x),max(x),0.01)
line_y=a*line_x+b

#그래프 그리기
plt.plot(line_x,line_y,'r-')
plt.plot(x,y,'bo')
plt.xlabel('Population Growth Rate (%)')
plt.ylabel('Elderly Population Rate (%)')
plt.show()
````



> 비선형 회귀

뉴런의 개수가 많으면 표현력이 좋아지지만 너무 많으면 학습이 제대로 안되거나 과적합 될 수 있다.
<br>
두 번째 레이어의 경우 x 입력값에 대한 하나의 y값을 출력하기에 뉴런 수는 1개이다.
<br><br>
  <font size="2em" color="gray">
  tanh : 시그모이드 함수와 개형이 유사, -1부터 1 사이 값 반환
  <br>
 과적합 : 학습 데이터에 대해서는 정확한 예측을 제공하지만 외부 데이터에 대해서는 그렇지 않은 경우
  </font>



````python
#모듈 불러오기
import tensorflow as tf
import numpy as np

#데이터 받기 (리스트)
x=[0.3,-0.78,1.26,0.03,1.11,0.24,-0.24,-0.47,-0.77,-0.37,-0.85,-0.41,-0.27,0.02,-0.76,2.66]
y=[12.27,14.44,11.87,18.75,17.52,16.37,19.78,19.51,12.65,14.74,10.72,21.94,12.83,15.51,17.14,14.42]

#모델 구성
model=tf.keras.Sequential([
   tf.keras.layers.Dense(units=6,activation='tanh',input_shape=(1,)),
   tf.keras.layers.Dense(units=1)
])

#모델 최적화 및 점검
model.compile(optimizer=tf.keras.optimizers.SGD(0.1),loss='mse') #손실 함수 : mse, 평균 제곱 오차
model.summary()
````

````python
# x를 입력하면 y가 정답이 되도록 10회 학습
model.fit(x,y,epochs=10)
# TIP : 손실값의 변화가 적으면 학습이 거의 다 된 것이다!
````

````python
#모듈 불러오기
import matplotlib.pyplot as plt

#회귀식 리스트로 
line_x=np.arange(min(x),max(x),0.01)
line_y=model.predict(line_x)

#그래프 그리기
plt.plot(line_x,line_y,'r-')
plt.plot(x,y,'bo')
plt.xlabel('Population Growth Rate (%)')
plt.ylabel('Elderly Population Rate (%)')
plt.show()
````
